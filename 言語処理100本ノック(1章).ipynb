00.文字列の逆順


```python
str = "stressed"
new_str = []

for count in range(0,len(str)):
    new_str.append(str[len(str)-count-1])

new_str = "".join(new_str) #リストを文字列に変換
print(new_str)
```

01. 「パタトクカシーー」

```python
str = "パタトクカシーー"
new_str = []

for i in [0,2,4,6]:
  new_str.append(str[i])

new_str = "".join(new_str)
print(new_str)
```

02. 「パトカー」＋「タクシー」＝「パタトクカシーー」

```python
str1 = "パトカー"
str2 = "タクシー"

new_str = []

for count in range(0,4):
  new_str.append(str1[count])
  new_str.append(str2[count])

new_str = "".join(new_str)
print(new_str)
```

03. 円周率

```python
str="Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics."
words = []
new_str = []
i = 0

for count in range(0,len(str)):
  if str[count] == " " or str[count] == ".":
    words.append(str[i:count])
    i = count+1

for count in range(0,len(words)):
  new_str.append(words[count][0])

print("".join(new_str))
```

04.元素記号

```python
str = "Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."
words = []
new_str = {}
i = 0

for count in range(0,len(str)):
  if str[count] == " " or str[count] == ".":
    words.append(str[i:count])
    i = count+1

words = [s for s in words if s != ''] #空文字の削除

for count in range (0, len(words)):
  if count in [0,4,5,6,7,8,14,15,18]:
    new_str[count] = words[count][0]
  else:
    new_str[count] = words[count][0:2]

print(new_str)
```
